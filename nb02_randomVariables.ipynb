In class we saw that a random variable $X$ that follows a  **geometric** distribution with parameter $p$ has the pmf 

$$p(X=k) = (1-p)^{k-1}p,$$

for $k=1,2,\ldots$ and $p(X=k)=0$ for all other $k$.
